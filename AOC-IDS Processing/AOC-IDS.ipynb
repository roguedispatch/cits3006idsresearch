# CICIDS2017 (Fixed)

```python
df_cicids_monday = pd.read_csv('../DATASETS/CICIDS2017_Fixed/FLOWS/Monday-WorkingHours.csv')
df_cicids_tuesday = pd.read_csv('../DATASETS/CICIDS2017_Fixed/FLOWS/Tuesday-WorkingHours.csv')
df_cicids_wednesday = pd.read_csv('../DATASETS/CICIDS2017_Fixed/FLOWS/Wednesday-workingHours.csv')
df_cicids_thursday = pd.read_csv('../DATASETS/CICIDS2017_Fixed/FLOWS/Thursday-WorkingHours.csv')
df_cicids_friday = pd.read_csv('../DATASETS/CICIDS2017_Fixed/FLOWS/Friday-WorkingHours.csv')

df_cicids = pd.concat([
    df_cicids_monday,
    df_cicids_tuesday,
    df_cicids_wednesday,
    df_cicids_thursday,
    df_cicids_friday,
])

df_cicids.rename(columns={'Label': 'label'}, inplace=True, errors='ignore')
df_cicids.loc[df_cicids['label'].str.contains("Attempted", case=False, na=False), 'label'] = "BENIGN"
df_cicids
```

```python
df_cicids['label'].value_counts(normalize=True)
```

```python
df_cicids.to_csv('../DATASETS/CICIDS2017_Fixed/FLOWS/combined.csv', index=False)
```

```python
df_cicids = df_cicids.drop([
    'Flow ID', 
    'Src IP', 
    'Src Port', 
    'Dst IP',
    'Dst Port', 
    'Timestamp',
], errors='ignore', axis=1)
```

```python
df_cicids
```

```python
# REMOVING INF

def drop_nonfinite_rows(x_df, y_df=None):
    numeric_cols = x_df.select_dtypes(include=[np.number]).columns
    mask = np.isfinite(x_df[numeric_cols]).all(axis=1)
    x_df_clean = x_df.loc[mask].copy()
    x_df_clean = x_df_clean.reset_index(drop=True)
    if y_df is not None:
        y_df_clean = y_df.loc[x_df.index[mask]].reset_index(drop=True)
        return x_df_clean, y_df_clean
    return x_df_clean

df_cicids = drop_nonfinite_rows(df_cicids)
```

```python
df_cicids.to_csv('../DATASETS/CICIDS2017_Fixed/FLOWS/AOC-IDS.csv', index=False)
```

# CTU13

```python
df_1 = pd.read_csv('../DATASETS/CTU13/CTU-13-Dataset/1/capture20110810.binetflow')
df_2 = pd.read_csv('../DATASETS/CTU13/CTU-13-Dataset/2/capture20110811.binetflow')
df_3 = pd.read_csv('../DATASETS/CTU13/CTU-13-Dataset/3/capture20110812.binetflow')
df_4 = pd.read_csv('../DATASETS/CTU13/CTU-13-Dataset/4/capture20110815.binetflow')
df_5 = pd.read_csv('../DATASETS/CTU13/CTU-13-Dataset/5/capture20110815-2.binetflow')
df_6 = pd.read_csv('../DATASETS/CTU13/CTU-13-Dataset/6/capture20110816.binetflow')
df_7 = pd.read_csv('../DATASETS/CTU13/CTU-13-Dataset/7/capture20110816-2.binetflow')
df_8 = pd.read_csv('../DATASETS/CTU13/CTU-13-Dataset/8/capture20110816-3.binetflow')
df_9 = pd.read_csv('../DATASETS/CTU13/CTU-13-Dataset/9/capture20110817.binetflow')
df_10 = pd.read_csv('../DATASETS/CTU13/CTU-13-Dataset/10/capture20110818.binetflow')
df_11 = pd.read_csv('../DATASETS/CTU13/CTU-13-Dataset/11/capture20110818-2.binetflow')
df_12 = pd.read_csv('../DATASETS/CTU13/CTU-13-Dataset/12/capture20110819.binetflow')
df_13 = pd.read_csv('../DATASETS/CTU13/CTU-13-Dataset/13/capture20110815-3.binetflow')

df_combined = pd.concat([
    df_1, df_2, df_3, df_4, df_5, df_6, df_7, df_8, df_9, df_10, df_11, df_12, df_13
])
```

```python
df_combined = df_combined.drop([
    'StartTime',
    'Dir'
], errors='ignore', axis=1)
df_combined.rename(columns={'Label': 'label'}, inplace=True, errors='ignore')
```

```python
df_combined['label'].value_counts()
```

```python
df_combined.isna().sum()
```

```python
df_combined = df_combined.dropna()
```

```python
# Filter botnet flows (those whose label contains "Botnet")
botnet_flows = df_combined[df_combined['label'].str.contains("Botnet", case=False, na=False)]

# Filter normal flows (assumed to be flows that do NOT contain "Botnet")
normal_flows = df_combined[df_combined['label'].str.contains("Normal", case=False, na=False)]
```

```python
botnet_flows['label'] = 1
normal_flows['label'] = 0
```

```python
df_final = pd.concat([botnet_flows, normal_flows])
```

```python
df_final.to_csv('../DATASETS/CTU13/FLOWS/AOC-IDS.csv', index=False)
```
